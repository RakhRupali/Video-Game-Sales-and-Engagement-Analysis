```python
games_df = games_df.dropna(subset=['Team'])
```

```python
#  Remove the opening and closing square brackets ---
# We use .str.replace() to find the literal brackets and replace them with an empty string ''.
games_df['Team'] = games_df['Team'].str.replace('[', '', regex=False)
games_df['Team'] = games_df['Team'].str.replace(']', '', regex=False)
games_df['Team'] = games_df['Team'].str.replace("'", '', regex=False)
games_df['Team'] = games_df['Team'].str.strip()
```

```python
games_df
```

```python
#  Remove the opening and closing square brackets ---
# We use .str.replace() to find the literal brackets and replace them with an empty string ''.
games_df['Reviews'] = games_df['Reviews'].str.replace('[', '', regex=False)
games_df['Reviews'] = games_df['Reviews'].str.replace(']', '', regex=False)
games_df['Reviews'] = games_df['Reviews'].str.replace("'", '', regex=False)
games_df['Reviews'] = games_df['Reviews'].str.strip()
```

```python
games_df
```

```python
games_df['Genres'] = games_df['Genres'].str.replace('[','',regex=False)
games_df['Genres'] = games_df['Genres'].str.replace(']','',regex=False)
games_df['Genres'] = games_df['Genres'].str.replace("'",'',regex=False)
games_df['Genres'] = games_df['Genres'].str.strip()
```

```python
games_df
```

```python
games_df.drop_duplicates(inplace=True) # removing duplicates value
```

# 3rd DATA FRAME merge data (games_df and vgsales_df)

```python
merged_df = pd.merge(
    left=games_df,
    right=sales_df,
    left_on='Title',      # Key column in games_df
    right_on='Name', # Key column in vgsales_df
    how='outer'           # Use 'outer' to keep all rows from both tables
)
```

```python
merged_df.dropna(inplace=True)
```
